```python
import tabula
import pandas as pd


pdf_path = "/content/sample_table_with_lines.pdf"


tables = tabula.read_pdf(pdf_path, pages='all', multiple_tables=True, lattice=True)


if tables:

    for idx, table in enumerate(tables):
        print(f"Table {idx + 1}:")
        print(table.head())


        table.to_csv(f"table_{idx + 1}.csv", index=False)
        print(f"Table {idx + 1} saved as table_{idx + 1}.csv")
else:
    print("No tables found in the PDF.")


combined_df = pd.concat(tables, ignore_index=True)
print("Combined DataFrame:")
print(combined_df)


combined_df.to_csv("combined_table.csv", index=False)
print("Combined table saved as combined_table.csv")
```
